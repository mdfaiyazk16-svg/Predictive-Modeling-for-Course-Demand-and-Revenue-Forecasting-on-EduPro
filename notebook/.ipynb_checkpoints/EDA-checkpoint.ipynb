### Imports

```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")

pd.set_option("display.max_columns", None)
```

### Load Dataset

```python
courses_df = pd.read_excel(
    "../data/raw/EduPro Online Platform.xlsx",
    sheet_name="Courses"
)

teachers_df = pd.read_excel(
    "../data/raw/EduPro Online Platform.xlsx",
    sheet_name="Teachers"
)

transactions_df = pd.read_excel(
    "../data/raw/EduPro Online Platform.xlsx",
    sheet_name="Transactions"
)

users_df = pd.read_excel(
    "../data/raw/EduPro Online Platform.xlsx",
    sheet_name="Users"
)

print(courses_df.shape)
print(teachers_df.shape)
print(transactions_df.shape)
print(users_df.shape)
```

### Dataset Overview

```python
display(courses_df.head())

display(teachers_df.head())

display(transactions_df.head())

display(users_df.head())
```

### Dataset Information

```python
print("\nCourses Info")
courses_df.info()

print("\nTeachers Info")
teachers_df.info()

print("\nTransactions Info")
transactions_df.info()

print("\nUsers Info")
users_df.info()
```

### Missing Values

```python
datasets = {
    "Courses": courses_df,
    "Teachers": teachers_df,
    "Transactions": transactions_df,
    "Users": users_df
}

for name, df in datasets.items():

    print(f"\n{name}")

    print(df.isnull().sum())
```

### Duplicate Analysis

```python
for name, df in datasets.items():

    print(
        f"{name} Duplicate Rows:",
        df.duplicated().sum()
    )
```

### Course Category Distribution

```python
plt.figure(figsize=(10,5))

sns.countplot(
    data=courses_df,
    x="CourseCategory"
)

plt.xticks(rotation=45)

plt.title(
    "Course Category Distribution"
)

plt.show()
```

### Monthly Transactions Trend

```python
transactions_df["TransactionDate"] = pd.to_datetime(
    transactions_df["TransactionDate"]
)

monthly = (
    transactions_df
    .groupby(
        transactions_df["TransactionDate"].dt.month
    )
    .size()
)

plt.figure(figsize=(10,5))

monthly.plot(marker="o")

plt.title(
    "Monthly Transaction Trend"
)

plt.ylabel("Transactions")

plt.show()
```

### Revenue Analysis

```python
revenue = (
    transactions_df
    .groupby("CourseID")["Amount"]
    .sum()
    .reset_index()
)

revenue = revenue.sort_values(
    "Amount",
    ascending=False
)

revenue.head(10)
```

### Top Revenue Courses

```python
top10 = revenue.head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    data=top10,
    x="CourseID",
    y="Amount"
)

plt.title(
    "Top 10 Revenue Generating Courses"
)

plt.show()
```

### Category Revenue Analysis

```python
merged = transactions_df.merge(
    courses_df,
    on="CourseID",
    how="left"
)

category_revenue = (
    merged.groupby(
        "CourseCategory"
    )["Amount"]
    .sum()
    .sort_values(
        ascending=False
    )
)

category_revenue
```

### Category Revenue Visualization

```python
plt.figure(figsize=(12,6))

category_revenue.plot(
    kind="bar"
)

plt.title(
    "Revenue by Category"
)

plt.ylabel("Revenue")

plt.show()
```

### Correlation Analysis

```python
numeric_cols = [

    "CoursePrice",
    "CourseDuration",
    "CourseRating"

]

corr_df = courses_df[numeric_cols]

plt.figure(figsize=(8,5))

sns.heatmap(
    corr_df.corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title(
    "Course Feature Correlation"
)

plt.show()
```
